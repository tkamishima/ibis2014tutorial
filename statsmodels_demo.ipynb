パッケージの読み込み

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
```

データの読み込み

```python
url = 'http://vincentarelbundock.github.io/Rdatasets/csv/HistData/Guerry.csv'
dat = pd.read_csv(url)
```

回帰モデルのあてはめ

```python
results = smf.ols('Lottery ~ Literacy + np.log(Pop1831)', data=dat).fit()
```

結果の表示

```python
print results.summary()
```
